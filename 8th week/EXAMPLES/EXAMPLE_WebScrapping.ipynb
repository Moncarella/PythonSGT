```python
from bs4 import BeautifulSoup

# soup = BeautifulSoup(response.content, "html.parser")

with open("European Central Bank.html","r",encoding="utf8") as f:
    data_html = f.read()
    
soup = BeautifulSoup(data_html,"html.parser")
```

```python
# with open("testpage.html", "w",encoding = "utf8") as f:
#     f.write(str(soup))
    
```

```python
inflation_element = soup.find(class_="ecb-inflationData cf")
```

```python
# inflation_elements = inflation_element.children   1st option how to get elements
inflation_elements = inflation_element.find_all(class_="item")
```

```python
# element = inflation_elements[0]
# element.text.split("%")[0]
```

```python
for inflation_data_element in inflation_elements:
    print("Inflation on " + inflation_data_element.find(class_="ecb-small").text +
          " is " + inflation_data_element.text.split("%")[0] + "%")
```

```python
# If we wanted to create a dictionary:
data = dict()

for inflation_data_element in inflation_elements:
    month = inflation_data_element.find(class_="ecb-small").text
    value = inflation_data_element.text.split("%")[0]
    data[month] = value
```

```python
# If we wanted to create a data frame:
import pandas as pnd
```

```python
df = pnd.DataFrame({"Month":data.keys(),"Inflation (%)":data.values()})
df
```

```python
with open("TestData.html","r",encoding="utf8") as f:
    html_content = f.read()
```

```python
soup = BeautifulSoup(html_content,"html.parser")
```

```python
table_html_element = soup.find("table")
```

```python
table_head = table_html_element.find("thead")
```

```python
# date_h = table_head.find_all("tr")[0].find("th").text   #Not an effective way
```

```python
html_header_rows = table_head.find_all("tr")
date_h = html_header_rows[0].find("th").text
```

```python
html_tab_columns = html_header_rows[1].find_all("th")
```

```python
columns_names = list()
for html_col in html_tab_columns:
    columns_names.append(html_col.text)
```

```python
# The data is stored under body
```

```python
rows_html = table_html_element.find("tbody").find_all("tr")
```

```python
row_names = list()
```

```python
rows = list()
```

```python
for row_html in rows_html:
    row_names.append(row_html.find("th").text.replace("..",""))
    columns_row = list()
    rows.append(columns_row)
    columns_html = row_html.find_all("td")
    for column_html in columns_html:
        columns_row.append(int(column_html.text.replace(",","").replace(" ","")))
```

```python
df = pnd.DataFrame(rows,columns=columns_names,index=row_names)
```

```python
df
```

```python
def get_age_group_data():
    age_group = input("Enter age group:\n")
    data = df.drop("TOTAL").drop("With income")[age_group]
    x = list(data.index)
    y = list(data)
    return [x,y,age_group]
```

```python
x,y,age_group = get_age_group_data()
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,label=age_group)
plt.xticks(range(len(x)),x,rotation=60)

plt.xlabel("Income")
plt.ylabel("Number of people")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
total = sum(y)
y = [100* yi/total for yi in y]

plt.plot(x,y,label=age_group)
plt.xticks(range(len(x)),x,rotation=60)

plt.xlabel("Income")
plt.ylabel("Number of people (%)")

plt.legend()
plt.show()
```

```python
x,y2,age_group2 = get_age_group_data()
total = sum(y2)
y2 = [100* yi/total for yi in y2]

plt.plot(x,y,label=age_group)
plt.plot(x,y2,label=age_group2)

plt.xticks(range(len(x)),x,rotation=90)

plt.xlabel("Income")
plt.ylabel("Number of people (%)")

plt.legend()
plt.show()
```
